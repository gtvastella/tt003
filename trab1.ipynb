Neste exercício, temos a lista de pessoas que estavam no Titanic com alguns dados de cada uma dessas pessoas. Os dados são apenas dados, podemos gerar alguma informação a partir deles?

## Importando Dados de Entrada



```python
## Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

## Importando o banco de dados com as listas de passageiros.
titanic = pd.read_csv('https://raw.github.com/mattdelhey/kaggle-titanic/master/Data/train.csv')

## Colocando os dados na tela
titanic
```

## Separando em Treino e Teste

```python
## Obtendo 80% dos dados para treino de forma aleatória
train = titanic.sample(frac=0.8,random_state=42)

## Colocando para teste tudo o que não foi usado no treino
test  = titanic.drop(train.index)

## Mostrando um resumo na tela.
print(train.describe(include = "all"))
print("----")
print(test.describe(include = "all"))
```

## Pré-processamento

Vamos olhar o banco de dados para decidir o nosso pré-processamento. Usaremos como fonte de informação o banco de dados de treino. Note com o comando a seguir que apenas **age**, **cabin** e **embarked** possuem *missing values*.

```python
train.info()
```

### Tratando Missing Values

De alguma forma, teremos que lidar com os missing values para os
 campos numéricos. Para **age**, vamos utilizar uma
 estratégia de preencher os dados que estão faltando com a mediana
 da coluna do banco de dados de treino em todos os casos.

 Para não contaminar os dados, **jamais use valores que estão no banco de dados de teste como fonte de informação para fazer esse pré-processamento**.

```python
## Preenchendo dados faltantes com a mediana do treino
medianAge = train.age.median()
train['age'] = train.age.fillna(medianAge)
test['age']  = test.age.fillna(medianAge)
```

No caso de **Cabin**, os missing values possuem uma semântica associada, são pessoas que não possuem cabines. Nesse caso, a falta de informação (missing values) é, por si só, uma informação. Vamos considerar apenas dois grupos de pessoas, as sem cabine e as com cabine.

```python
train["cabin"] = train.cabin.notna().astype(int)
test["cabin"]  = test.cabin.notna().astype(int)
```

No caso de embarked, temos dois registros nulos apenas. Como grande parte das pessoas embarcaram em Southampton, iremos dar um "chute" e assumir que esses dois passageiros vieram de lá. Isso possivelmente inclui ruído no modelo. Uma outra abordagem seria eliminar os registros. Como são registros não confiáveis e são bem poucos, qualquer uma das abordagens seria válida.

```python
## Código para substituir NaN por S
train["embarked"] = train.embarked.fillna("S")
test["embarked"] = test.embarked.fillna("S")
```

Agora que fizemos modificações, devemos checar novamente como ficou
o banco de dados. Note que parece melhor do que antes, dado não termos mais **Missing Values**.

```python
## Estamos combinando o banco de dados novamente apenas para
## ver se resolvemos todos os problemas em ambos os casos.
pd.concat([train, test]).info()
```

Note na saída do comando acima que **name**, **sex**, **ticket** e **embarked** estão marcados com *Dtype* do tipo *object*. Isso quer dizer que não são numéricos, mas a entrada da rede neural tem que ser sempre numérica.

Assuma que queremos usar **sex** e **embarked** na análise e ignorar os outros dois, então vamos ter que transformar os valores dessas colunas para servirem como entrada.

## Gerando valores numéricos

Temos
que mudar os valores que estão em **"Sex"** e em **"Embarked"** para números. No caso de **sex**, transformaremos **female** em 1 e **male** em 0.

```python
train["sex"] = (train.sex == "female").astype(int)
test["sex"]  = (test.sex  == "female").astype(int)
```

Para **Embarked**, teremos um pouco mais de trabalho. Poderíamos colocar "S", "Q" e "C" com os números 0, 1 e 2, por exemplo. Entretanto, as redes neurais podem achar que "C" é duas vezes maior do que "Q", por exemplo, e tentar fazer alguma suposição a partir disso, o que não faria sentido. Portanto, usaremos uma estratégia conhecida como OneHotEncoder. Primeiro, veja o código a seguir:

```python
pd.get_dummies(train.embarked)
```

O código obteve os valores (C, Q ou S) da coluna **embarked** e os tranformou em três colunas: **C**, **Q** e **S**, atribuindo 1 se anteriormente o valor em embarked era o valor da coluna, e 0 caso contrário.

O que faremos agora é inserir as novas colunas **C**, **Q** e **S** nos dataframes.

```python
train = pd.concat([train, pd.get_dummies(train.embarked).astype(int)], axis = "columns")
test =  pd.concat([test,  pd.get_dummies(test.embarked).astype(int)],  axis = "columns")
```

Vamos olhar agora os dados.

```python
train
```

A nossa coluna **pclass** já tem dados numéricos. Entretanto, como argumentado antes, algum modelo de aprendizado de máquina pode interpretar que alguém da classe 3 é três vezes maior (em algum sentido) que alguém da classe 1. Entretanto, essa não é a semântica que queremos associar a esses valores. Por isso, vamos mudar os valores para OneHotEncoding. Antes de mudar os valores usando o procedimento que fizemos para Embarked, iremos transformar os valores numéricos em strings da seguinte forma:

```python
train["pclass"] = train.pclass.map(lambda x : "classe_"+str(x))
test["pclass"]  = test.pclass.map(lambda x : "classe_"+str(x))
```

```python
train
```

Agora faremos a transformação:

```python
train = pd.concat([train, pd.get_dummies(train.pclass).astype(int)], axis = "columns")
test =  pd.concat([test,  pd.get_dummies(test.pclass).astype(int)],  axis = "columns")
```

Temos agora um conjunto de dados com várias colunas tratadas. Essas colunas podem ser usadas para treinar uma rede neural.

## Organizando os Dados

Vamos agora gerar os nossos dados para treinar a rede neural.

```python
x_train = train.loc[: , ["sex", "age", "sibsp", 'parch','fare','cabin', 'C', 'Q', 'S', 'classe_1', 'classe_2', 'classe_3']]
y_train = train.survived
x_test  = test.loc[:, ["sex", "age", "sibsp", 'parch','fare','cabin', 'C', 'Q', 'S', 'classe_1', 'classe_2', 'classe_3']]
y_test  = test.survived
```

O modo como você irá gerar os tensores dependerá de qual a função de custo que você irá utilizar. Você tem duas opções. Escolha a que achar mais adequada.

```python
import torch
```

```python
## Opção 1:
x_train_op1 = torch.tensor(x_train.values, dtype = torch.float)
x_test_op1  = torch.tensor(x_test.values,  dtype = torch.float)
y_train_op1 = torch.tensor(y_train.values, dtype = torch.float)
y_test_op1  = torch.tensor(y_test.values,  dtype = torch.float)
```

```python
## Opção 2:
x_train_op2 = torch.tensor(x_train.values, dtype = torch.float)
x_test_op2  = torch.tensor(x_test.values,  dtype = torch.float)
y_train_op2 = torch.tensor(y_train.values, dtype = torch.long)
y_test_op2  = torch.tensor(y_test.values,  dtype = torch.long)
```

```python
xt = x_train_op1
yt = y_train_op1.unsqueeze(-1)

xf = x_test_op1
yf = y_test_op1.unsqueeze(-1)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
class BinaryClassifier(nn.Module):
    def __init__(self, input_dim):
        super(BinaryClassifier, self).__init__()
        self.input_layer = nn.Linear(input_dim, 128)
        self.hidden_layer = nn.Linear(128, 64)
        self.output_layer = nn.Linear(64, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        out = torch.relu(self.input_layer(x))
        out = torch.relu(self.hidden_layer(out))
        out = self.sigmoid(self.output_layer(out))
        return out
```

```python
input_dim = 12
model = BinaryClassifier(input_dim)
```

```python
loss_fn = torch.nn.BCELoss()
learning_rate = 0.001
optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.65)
epoches = 500
minib = 64
```

```python
for epoch in range(epoches):

    for i in range(0, len(xt), minib):
        x_batch = xt[i:i+minib]
        y_batch = yt[i:i+minib]


        optimizer.zero_grad()


        y_pred = model(x_batch)


        loss = loss_fn(y_pred, y_batch)


        loss.backward()
        optimizer.step()


    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{epoches}], Loss: {loss.item()}')
```

```python



predictionst = model(xt)
predictionsf = model(xf)


survivedt = (predictionst > 0.5).float()
survivedf = (predictionsf > 0.5).float()


accuracytrain = (survivedt == yt).float().mean().item()
accuracytest = (survivedf == yf).float().mean().item()

print(f'Acurácia treino: {accuracytrain * 100:.2f}%')
print(f'Acurácia teste: {accuracytest * 100:.2f}%')
```
